```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
import requests  # unchanged

BASE_URL = "https://www.playfantasyrugby.com/json/fantasy" 

def download_file(filename):
    response = requests.get(f"{BASE_URL}/{filename}")

    with open(f"../data/raw/{filename}", "wb") as file:  
        file.write(response.content)  


download_file("squads.json")
download_file("player_stats.json")
download_file("players.json")
download_file("rounds.json")
```

```python
squads = pd.read_json("../data/raw/squads.json")
squads.to_csv("../data/processed/squads.csv", index=False)
```

```python
player_stats = pd.read_json("../data/raw/player_stats.json", orient="index").sort_index().reset_index().rename(columns={'index':'id'})
player_stats.to_csv("../data/processed/player_stats.csv", index=False)
```

```python
players_raw = pd.read_json("../data/raw/players.json")

players = players_raw.drop(columns=['stats','imagePitch','imageProfile','selected'])#.query("status != 'eliminated' ")

players = players.join(pd.json_normalize(players_raw.stats))
players = players.join(pd.json_normalize(players_raw.selected))

players.to_csv("../data/processed/players.csv", index=False)
```

```python
rounds = pd.read_json("../data/raw/rounds.json")
rounds
```

```python
df = pd.DataFrame()
for row in rounds.itertuples():
    df = pd.concat([df, pd.json_normalize(row.tournaments)])
```

```python
df.to_csv("../data/processed/rounds.csv", index=False)
```
