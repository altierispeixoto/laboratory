```python
players = pd.read_csv("../data/processed/players.csv")

player_stats = pd.read_csv("../data/processed/player_stats.csv")
```

```python
players = players.merge(player_stats, on='id', how='inner').query("status != 'eliminated' ") #and status != 'injured'
```

```python
positions = {
                'prop':'prop',
                'hooker':'hooker', 
                'lock': 'lock',
                'loose_forward': 'loose_forward', 
                'fly_half':'fly_half',
                'center':'midfield',
                'scrum_half': 'half_back',
                'outside_back':'back_three',
            }

players['position'] = players['position'].map(positions)
```

```python
import pandas as pd
from mip import Model, xsum, maximize, BINARY

# Constants
BUDGET = 100_000_000
MAX_PLAYERS_TEAM = 4

positions = {
    'prop': 2,
    'hooker': 1,
    'lock': 2,
    'loose_forward': 3,
    'half_back': 1,
    'fly_half': 1,
    'midfield': 2,
    'back_three': 3
}

# Load player data (assuming you have a CSV file with player information)
df = players.copy()

# Create the model
model = Model()

# Create binary variables for each player
x = [model.add_var(var_type=BINARY) for i in range(len(df))]

# Objective function: Maximize total player rating
model.objective = maximize(xsum(df.iloc[i]['totalPoints'] * x[i] for i in range(len(df))))

# Budget constraint
model += xsum(df.iloc[i]['cost'] * x[i] for i in range(len(df))) <= BUDGET

# Position constraints
for position, count in positions.items():
    model += xsum(x[i] for i in range(len(df)) if df.iloc[i]['position'] == position) == count

# Team constraint
for team in df['squadId'].unique():
    model += xsum(x[i] for i in range(len(df)) if df.iloc[i]['squadId'] == team) <= MAX_PLAYERS_TEAM

# Solve the model
model.optimize()

# Get the selected players
selected_players = df[pd.Series(x, index=df.index).apply(lambda v: v.x > 0.5)]

# Print the results
print("Selected Players:")
#print(selected_players[['firstName','lastName', 'position', 'squadId', 'cost', 'totalPoints']])
print(f"\nTotal Budget Used: ${selected_players['cost'].sum():,}")
print(f"Total Rating: {selected_players['totalPoints'].sum():.2f}")
```
